## Ejercisio 1: Estructuras Básicas de Datos con Python

```python
familia = {                                 
    "Elkin": (29, 65, 1.70, "No estudia"),  
    "Fernando": (27, 62, 1.75, "Estudia"),  
    "Arturo": (70, 78, 1.60, "No estudia"), 
    "Angelica": (35, 85, 1.78, "Estudia"),  
    "Biatriz": (29, 62, 1.68, "Estudia"),
    "Ignacio": (15, 35, 1.55, "Estudia"),
    "Ainoa": (5, 20, 1.35, "Estudia")
}


# Creamos una variable vacia la cual almacenara los valores de la lista. 

edades = []
pesos = []
estaturas = []
estudia = []

# ciclo for 

for key, i in familia.items():        
    edades.append(i[0])               
    pesos.append(i[1])               
    estaturas.append(i[2])            
    

# Mostrar los resultados
print("Edades:", edades)
print("Pesos:", pesos)
print("Estaturas:", estaturas)
print("Estudia:", estudia)
```

```python
# Funcion que devuelve el promedio

def promedio(lista):            
    
    suma=0                      
    contador=0                  
    for i in range(len(lista)): 
            contador+=1         
            suma+=lista[i]      
            
    return(suma/contador)       
    print(promedio(lista))      



print(" La edad promedio de la familia de Fernando es:",promedio(edades))
print(" El peso promedio de la familia de Fernando es:",promedio(pesos))
print(" La estarura promedio de la familia de Fernando es:",promedio(estaturas))

```

```python
# Cargar paquetes
from sklearn.datasets import load_breast_cancer


data=load_breast_cancer() # Nos arroja un diccionario de los datos por lo que lo convertimos a un dataframe de pandas para mejorar la manipulación

df3= pd.DataFrame(data.data, columns=data.feature_names)

```

utilizamos el siguiente codigo de indexacion y igualacion. 

```python
df3['target'] = data.target # Agrega a la variable df3 una colummna llamada target la cual Contiene las etiquetas de clase (0 = maligno, 1 = benigno) 
df3.head()
```

```python
# Construcción del panel 

#  diccionario 
panel = {
    "Conjunto_de_datos_1": df1,
    "Conjunto_de_datos_2": df2,
    "breast_cancer": df3
}

panel
```

```python
# funcion para imprimir la cantidad de elementos del panel

def cantidad(panel):
    for i, df in panel.items(): 
        num_elementos = df.size # Para obtener  el tamaño de cada DataFrame 
        print(f'La base de datos "{i}" tiene {num_elementos} elementos.')

# Llamada a la función con el Panel
cantidad(panel)
```

## Ejercisio 3: Gráficos científicos

```python
import numpy as np
import matplotlib.pyplot as plt

# Puntos de dispersión: Se crean directamente usando plt.scatter
plt.scatter([13, 14, 15, 16, 17, 18], [16, 24, 45, 57, 75, 102], color='red', label='Puntos de dispersión')

```

```python
# Rango de x para las funciones
x = np.linspace(13, 18)

# Funciones en el mismo gráfico
plt.plot(x, 17 * x - 210.3333, color='blue', label='Función lineal f(x) = 17x - 210.3333')
plt.plot(x, 0.1452 * np.exp(0.369 * x), color='green', label='Función exponencial g(x) = 0.1452e^(0.369x)')
plt.plot(x, 259.49 * np.log(x) - 656.46, color='red', label='Función logarítmica h(x) = 259.49ln(x) - 656.46')

# Detalles del gráfico
plt.xlabel("Tiempo") # Nombre del eje X
plt.ylabel("Resultado") # Nombre del eje Y
plt.title("Gráfica de Puntos y Funciones") # Titulo del gráfico 
plt.legend() # Mostrar las leyendas


# Mostrar la gráfica
plt.show()
```

**Conclusiones de la gráfica segun los nombres de los ejes que seleccione**

De acuerdo con la gráfica, se puede observar que a medida que el tiempo avanza, el resultado muestra una tendencia creciente. Esto refleja una relación donde el tiempo impulsa un aumento en el resultado, aunque el ritmo de crecimiento puede variar dependiendo de la función que mejor ajuste a los datos observados.Especificamente si se quiere un resultado mayor se puede usar una función exponencial que es la que llega a un pico más alto. 

```python
# Primero vamos a indenxar las variables que necesitamos en especifico, es decir, seleccionar de la base de datos df1
# solamente los atributos, tiempo en motor y tiempo en movimiento.Esto mediante corchetes como se muestra a continuación.

Tiempo_motor=df1["Tiempo motor ON (h)"]
Tiempo_en_movimiento=df1["Tiempo en Movimiento (h)"]
```

Para crear el histograma, usamos la funcion hist del paquete plt, alpha para controlar el nivel de transparencia, bins para especificar el numero de intervalos del histograma, en esta ocación lo dejaremos en defecto.

```python
plt.figure(figsize=(8, 6))

plt.hist(Tiempo_motor,alpha=0.5, color='blue', label='Tiempo motor')  

plt.hist(Tiempo_en_movimiento,alpha=0.5, color='green', label='Tiempo en movimiento') 

plt.legend()

plt.xlabel('Valores')

plt.ylabel('Frecuencia')

plt.title('Histograma de frecuencia')


```

```python

plt.scatter(Tiempo_motor,Tiempo_en_movimiento)


plt.xlabel('Valores')

plt.ylabel('y')

plt.title("Dispersion de los datos")
```

Analisis y conclusion de las graficas 

De acuerdo con el histograma de frecuencias se puede observar que a mayor tiempo en motor, los valores tienden a extenderse hacia un rango superior en comparación con el tiempo en movimiento. Esto sugiere que el motor suele estar activo durante periodos más largos que el tiempo en el que realmente hay movimiento.

Por otra parte de acuerdo al grafico de dispersion los puntos siguen un patrón alineado, lo cual sugiere que hay una relación lineal directa entre las dos variables. Esto indica que, conforme aumentan el tiempo del motor , el tiempo del movimiento también incrementa de manera casi proporcional.
